```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use("ggplot")
```

```python
x_data = np.random.random(50)*100
y_data = np.random.random(50) * 100
```

```python
plt.scatter(x_data,y_data,c='purple',s=120,alpha=0.5)
```

```python
years = [2006 + x for x in range(16)]
weights = [180 - x for x in range(16)]
plt.plot(years,weights,c='r',lw=3,linestyle='--')
```

```python
x = ['C++','C#','Python','Java','Go']
y = [20,50,140,1,45]
plt.bar(x,y,color='green',align='edge',width=0.9)
```

```python
ages = np.random.normal(20,1.5,1000)
plt.hist(ages,bins=20,color='blue',cumulative=True)
```

```python
explodes = [0.2,0,0,0,0]
plt.pie(y,labels=x,explode=explodes,autopct="%.2f%%")
```

```python
heights = np.random.normal(172,8,300)
plt.boxplot(heights)
```

```python
years = list(range(2014,2022))
income = [55,56,62,61,72,72,73,75]
income_ticks = list(range(50,81,2))

plt.title("Jhon's income (USD)",fontsize=20,fontname='century gothic')
plt.xlabel('Years')
plt.ylabel("Yearly Income in USD")
plt.yticks(income_ticks,[f'{x}k USD' for x in income_ticks])
plt.plot(years,income)
plt.show()
```

```python
values = [10,40,30,24,9]
labels = ['Java',"Python",'Brainfuck','JavaScript','C']

plt.pie(values,labels=None)
plt.legend(labels = labels)
```

```python
comp1 = list(range(45,80))
comp2 = list(range(30,70))

plt.plot(comp1,label='Tesla')
plt.plot(comp2,label='Facebook')
plt.legend()
```
