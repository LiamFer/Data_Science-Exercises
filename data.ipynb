```python
class nWalk():
    
    def __init__(self,num_points=5000):
        self.num_points = num_points
        self.x_values = [0]
        self.y_values = [0]
    
    def get_step(self):
        direction = choice([1,-1])
        distance = choice(list(range(1,9)))
        return direction * distance
        
    def fill_walk(self):
        while len(self.x_values) < self.num_points:
            x_step = self.get_step()

            y_step = self.get_step()
            
            if x_step == 0 and y_step == 0:
                continue
            
            nextx = self.x_values[-1] + x_step
            nexty = self.y_values[-1] + y_step
            
            self.x_values.append(nextx)
            self.y_values.append(nexty)
            
rw = nWalk(500000)
rw.fill_walk()

plt.scatter(rw.x_values,rw.y_values,s=1,c=list(range(rw.num_points)),cmap=plt.cm.Purples)
plt.scatter(0,0,c='green',s=100)
plt.scatter(rw.x_values[-1],rw.y_values[-1],c='red',s=100)
plt.show()
```

```python

rw = nWalk()
rw.fill_walk()

plt.plot(rw.x_values,rw.y_values,linewidth=1,c="red")
plt.show()
```

```python
class dice():
    def __init__(self,sides:int):
        self.sides = sides
    
    def roll(self):
        return choice(list(range(1,self.sides+1)))

results = []
heights = []
dado = dice(6)
for k in range(100):
    results.append(dado.roll())
    
    
for side in list(range(1,dado.sides+1)):
    heights.append(results.count(side))

print(heights)
print(sum(heights))

hist = pygal.Bar()

hist.title = "Results of rolling one D6 1000 times."
hist.x_labels = ['1', '2', '3', '4', '5', '6']
hist.x_title = "Result"
hist.y_title = "Frequency of Result"
hist.add('D6', heights)
hist.render_to_file('die_visual.svg')

plt.bar(list(range(1,dado.sides+1)),heights)
plt.show()
        
```
