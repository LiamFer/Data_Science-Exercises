### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
tips
colors = list(map(lambda k: "blue" if k[0]=='M' else "pink",tips.sex))
plt.scatter(tips.tip,tips.day,c=colors,alpha=0.5)
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python


fig,axs = plt.subplots(1,2,figsize=(9,4))
axs[0].hist(tips[tips.time=='Dinner'].tip)
axs[0].set(title="Histogram - Dinner Tip",xlabel="Tip",ylabel='Count')
axs[1].hist(tips[tips.time!='Dinner'].tip,color='lightgreen')
axs[1].set(title="Histogram - Lunch Tip",xlabel="Tip",ylabel='Count')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
fig,axs = plt.subplots(1,2,figsize=(9,4))

axs[0].scatter(tips[tips.sex=='Female'].tip,tips[tips.sex=='Female'].total_bill,c=list(map(lambda x: "lightgreen" if x[0]=='Y' else "red",tips[tips.sex=='Female'].smoker)))
axs[0].set(title='Male')

axs[1].scatter(tips[tips.sex!='Female'].tip,tips[tips.sex!='Female'].total_bill,c=list(map(lambda x: "lightgreen" if x[0]=='Y' else "red",tips[tips.sex!='Female'].smoker)))
axs[1].set(title='Female')

plt.tight_layout()
```
