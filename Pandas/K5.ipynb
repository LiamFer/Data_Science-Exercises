```python
import pandas as pd
```

```python
ufo = pd.read_csv(r"http://bit.ly/uforeports")
```

```python
ufo.head(3)
```

```python
# loc é para labels
# loc = loc[rows,columns] sendo que : se refere a todas as colunas
# ou todas as linhas
ufo.loc[0,:]
# Pra selecionar as linhas de 1 até 4 podemos fazer 0:3
# Pra selecionar da coluna city até state podemos fazer city:state
ufo.loc[0:3,:]
```

```python
#dá pra dar query usando o .loc na linha
ufo.loc[ufo['City']=="Oakland",:]
```

```python
# O .iloc usa como base a posição em int
ufo.iloc[:,0:4]
```
